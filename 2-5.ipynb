2. Time this function for various n e.g. n = 1,2,3.... You should have small values of n all the way up to large values. Plot "time" vs "n" (time on y-axis and n on x-axis). Also, fit a curve to your data, hint it's a polynomial. 

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import time

def f(n):
    x = 1
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            x += 1
    return x


n_values = np.arange(1, 1001)


execution_times = []
for n in n_values:
    start_time = time.time()
    f(n)
    end_time = time.time()
    execution_times.append(end_time - start_time)


def polynomial_func(x, a, b, c):
    return a * x**2 + b * x + c

popt, _ = curve_fit(polynomial_func, n_values, execution_times)


plt.plot(n_values, execution_times, label='Data', color='blue')
plt.plot(n_values, polynomial_func(n_values, *popt), label='Curve', color='red')
plt.xlabel('n')
plt.ylabel('Time (seconds)')
plt.title('Time vs n')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import time

def modified_f(n):
    x = 1
    y = 1
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            x += 1
            y = i + j
    return x


n_values = np.arange(1, 1001)


modified_execution_times = []
for n in n_values:
    start_time = time.time()
    modified_f(n)
    end_time = time.time()
    modified_execution_times.append(end_time - start_time)


def polynomial_func(x, a, b, c):
    return a * x**2 + b * x + c

popt, _ = curve_fit(polynomial_func, n_values, modified_execution_times)


plt.plot(n_values, modified_execution_times, label='Modified Data', color='blue')
plt.plot(n_values, polynomial_func(n_values, *popt), label='Curve', color='red')
plt.xlabel('n')
plt.ylabel('Time (seconds)')
plt.title('n')
plt.legend()
plt.grid(True)
plt.show()
```
